```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer_sur = pd.read_csv('haberman.csv')
# Display the first few rows of the dataset
cancer_sur.head()
```

```python
# add column names
cancer_sur.columns = ['age', 'year_of_operation', 'nodes', 'survived']
# Display the first few rows of the dataset with new column names
cancer_sur.head()
```

```python
# Display the summary statistics of the dataset
cancer_sur.describe()
```

```python
sns.FacetGrid(cancer_sur, hue="survived", height=5).map(sns.distplot, "age").add_legend()
```

```python
sns.FacetGrid(cancer_sur, hue="survived", height=5).map(sns.distplot, "year_of_operation").add_legend()
```

```python

counts, bin_edges = np.histogram(cancer_sur['age'], bins=10, density=True)

plt.xlabel('age')

pdf = counts / sum(counts)
print("pdf =", pdf)
print("bin_edges =", bin_edges)

cdf = np.cumsum(pdf)
print("cdf =", cdf)

plt.plot(bin_edges[1:], pdf, label='PDF')
plt.plot(bin_edges[1:], cdf, label='CDF')
plt.legend()
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=cancer_sur, hue='survived', palette='Set3')
plt.show()
```

```python
sns.boxplot(x='survived', y='year_of_operation', data=cancer_sur, hue='survived', palette='Set3')
plt.show()
```

```python
sns.violinplot(x="survived", y="year_of_operation",data=cancer_sur,width=0.8, hue='survived', palette='Set3')
plt.show()
```

```python
sns.pairplot(cancer_sur,hue="survived",height=3)
plt.show()
```
